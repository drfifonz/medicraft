```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 400)

# Define the ReLU function
relu = np.maximum(0, x)

# Create the plot
plt.plot(x, relu, label="ReLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funkcja aktywacji ReLU f(x) = max(0, x)")
plt.legend()
# plt.grid(True)
plt.show()
```
